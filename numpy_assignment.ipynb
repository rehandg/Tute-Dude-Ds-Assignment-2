# Question 1

In this section, a 1D NumPy array containing integers from 1 to 20 is created. Statistical operations such as sum, mean, median, and standard deviation are performed.


```python
import numpy as np

# 1D Array
arr1 = np.arange(1, 21)

print("1D Array:")
print(arr1)

print("\nSum:", np.sum(arr1))
print("Mean:", np.mean(arr1))
print("Median:", np.median(arr1))
print("Standard Deviation:", np.std(arr1))

indices = np.where(arr1 > 10)
print("Indices of elements greater than 10:", indices)

# 2D Array
arr2 = np.arange(1, 17).reshape(4, 4)

print("\n4x4 Array:")
print(arr2)

print("\nTranspose:")
print(arr2.T)

print("\nRow-wise Sum:")
print(np.sum(arr2, axis=1))

print("\nColumn-wise Sum:")
print(np.sum(arr2, axis=0))

# Random 3x3 Arrays
a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))

print("\nArray A:")
print(a)

print("\nArray B:")
print(b)

print("\nAddition:")
print(a + b)

print("\nSubtraction:")
print(a - b)

print("\nMultiplication:")
print(a * b)

print("\nDot Product:")
print(np.dot(a, b))

# Reshape and Slice
arr3 = np.arange(1, 13).reshape(3, 4)

print("\n3x4 Array:")
print(arr3)

print("\nFirst Two Rows and Last Two Columns:")
print(arr3[:2, -2:])
```
