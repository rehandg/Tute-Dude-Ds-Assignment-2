# Question 1

A DataFrame is created with employee details. Operations such as displaying rows, summary statistics, and calculating average salary are performed.

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)

print("First Five Rows:")
print(df.head())

print("\nSummary Statistics:")
print(df[['Age', 'Salary']].describe())

hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("\nAverage Salary of HR Department:", hr_avg_salary)

# Add Bonus Column
df['Bonus'] = df['Salary'] * 0.10

print("\nDataFrame with Bonus Column:")
print(df)

# Filter Employees Age Between 25 and 30
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

print("\nEmployees Aged Between 25 and 30:")
print(filtered_df)

# Group by Department
grouped = df.groupby('Department')['Salary'].mean()

print("\nAverage Salary by Department:")
print(grouped)

# Sort by Salary
sorted_df = df.sort_values(by='Salary')

print("\nSorted DataFrame:")
print(sorted_df)

sorted_df.to_csv("sorted_employees.csv", index=False)

print("\nCSV file saved successfully.")
```
