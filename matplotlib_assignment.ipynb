# Question 1 - Line Plot

A simple line plot is created using x and y values. The graph includes a title, labels, and grid.

```python
import matplotlib.pyplot as plt
import numpy as np

# Line Plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()

# Bar Graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# Pie Chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1, 0, 0, 0]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode)
plt.title("Revenue Distribution")
plt.show()

# Histogram
random_numbers = np.random.randint(1, 101, 1000)

plt.hist(random_numbers, bins=10)
plt.title("Histogram of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
